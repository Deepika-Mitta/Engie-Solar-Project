```python
query = ''' WITH CTE as (SELECT * FROM df_final_bb_cons)
SELECT *, sum(GHI) over(partition by Minute) as Hourly_GHI, 
sum(Energy) over(partition by Minute) as Hourly_Energy FROM CTE GROUP by Year, Month, Day, Hour, Minute, Temperature, "Clearsky DHI",
       "Clearsky DNI", "Clearsky GHI", "Cloud Type", "Dew Point", DHI, DNI,
       "Fill Flag", GHI, "Relative Humidity", "Solar Zenith Angle",
       "Surface Albedo", Pressure, "Precipitable Water", "Wind Direction",
       "Wind Speed", "Global Horizontal UV Irradiance (280-400nm)",
       "Global Horizontal UV Irradiance (295-385nm)"
'''
```

```python
query_out = sqldf(query)
```

```python
query_out.head(50)
```

```python
query_out['conversion_efficiency'] = query_out['Hourly_Energy']*1000/(query_out['Hourly_GHI']*24.8171891)
```

```python
sorted_df_hourly = query_out.sort_values(by = 'Hour')
```

```python
selected_columns_df = sorted_df_hourly[['Year','Month', 'Day','Hour','GHI','Energy','conversion_efficiency']]

# Specify the path where you want to save the CSV file

# Write the selected columns to the CSV file
selected_columns_df.to_csv('Hourly_eff.csv')
```

```python
query_Daily = ''' WITH CTE as (SELECT * FROM df_final_bb_cons)
SELECT *, sum(GHI) over(partition by Hour) as Daily_GHI, 
sum(Energy) over(partition by Hour) as Daily_Energy FROM CTE GROUP by Year, Month, Day, Hour, Minute, Temperature, "Clearsky DHI",
       "Clearsky DNI", "Clearsky GHI", "Cloud Type", "Dew Point", DHI, DNI,
       "Fill Flag", GHI, "Relative Humidity", "Solar Zenith Angle",
       "Surface Albedo", Pressure, "Precipitable Water", "Wind Direction",
       "Wind Speed", "Global Horizontal UV Irradiance (280-400nm)",
       "Global Horizontal UV Irradiance (295-385nm)"
'''
```

```python
Daily_df = sqldf(query_Daily)
```

```python
Daily_df['conversion_efficiency'] = Daily_df['Daily_Energy']*1000/(Daily_df['Daily_GHI']*24.8171891)
```

```python
sorted_daily_df= Daily_df.sort_values(by = 'Day')
```
