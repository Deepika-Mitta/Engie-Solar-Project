```python
query = ''' WITH CTE as (SELECT * FROM df_final_ev_cons)
SELECT *, sum(GHI) over(partition by Minute) as Hourly_GHI, 
sum(Energy) over(partition by Minute) as Hourly_Energy FROM CTE GROUP by Year, Month, Day, Hour, Minute, Temperature, "Clearsky DHI",
       "Clearsky DNI", "Clearsky GHI", "Cloud Type", "Dew Point", DHI, DNI,
       "Fill Flag", GHI, "Relative Humidity", "Solar Zenith Angle",
       "Surface Albedo", Pressure, "Precipitable Water", "Wind Direction",
       "Wind Speed", "Global Horizontal UV Irradiance (280-400nm)",
       "Global Horizontal UV Irradiance (295-385nm)"
'''
```

```python
EV_hourly = sqldf(query)
```

```python
EV_hourly.head()
```

```python
EV_hourly['conversion_efficiency'] = EV_hourly['Hourly_Energy']*1000/(EV_hourly['Hourly_GHI']*16.7225)
```

```python
sorted_EV_hourly = EV_hourly.sort_values(by = 'Hour')
```

```python
selected_columns_df = sorted_EV_hourly[['Year','Month', 'Day','Hour','GHI','Energy','conversion_efficiency']]

# Specify the path where you want to save the CSV file

# Write the selected columns to the CSV file
selected_columns_df.to_csv('EV_Hourly_eff.csv')
```

```python
query_Daily = ''' WITH CTE as (SELECT * FROM df_final_ev_cons)
SELECT *, sum(GHI) over(partition by Hour) as Daily_GHI, 
sum(Energy) over(partition by Hour) as Daily_Energy FROM CTE GROUP by Year, Month, Day, Hour, Minute, Temperature, "Clearsky DHI",
       "Clearsky DNI", "Clearsky GHI", "Cloud Type", "Dew Point", DHI, DNI,
       "Fill Flag", GHI, "Relative Humidity", "Solar Zenith Angle",
       "Surface Albedo", Pressure, "Precipitable Water", "Wind Direction",
       "Wind Speed", "Global Horizontal UV Irradiance (280-400nm)",
       "Global Horizontal UV Irradiance (295-385nm)"
'''
```

```python
EV_Daily_df = sqldf(query_Daily)
```

```python
EV_Daily_df['conversion_efficiency'] = EV_Daily_df['Daily_Energy']*1000/(EV_Daily_df['Daily_GHI']*16.7225)
```

```python
EV_sorted_daily_df= EV_Daily_df.sort_values(by = 'Day')
```

```python

# Create a new DataFrame with only the selected columns
selected_columns_df = EV_sorted_daily_df[['Year','Month', 'Day','GHI','Energy','conversion_efficiency']]

# Specify the path where you want to save the CSV file

# Write the selected columns to the CSV file
selected_columns_df.to_csv('EV_Daily_eff.csv')
```
